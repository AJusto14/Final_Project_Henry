```python
def calculate_percent_null(df, filter= False, percent=60.0):
  '''
  Función que devuelve un dataframe con todas las columnas del dataframe del parametro
  con el porcentaje de nulos que tiene cada columna.

  Con el parametro de "filter=True", la función devuelve al dataframe pasado como parametro con las columnas que tienen
  un porcentaje de nulos menor al parametro de "percent", que por default es igual a 60.0 (60%)

  '''
  if filter == False:
    return pd.DataFrame([(column, (df[column].isna().sum() * 100) / len(df)) for column in df.columns],
                        columns=['Column','% of Null Values']).sort_values(by='% of Null Values', ascending=False).reset_index(drop=True)
  elif filter == True:
    columns = []
    for c in df.columns:
      if df[c].isna().sum() * 100 / len(df) < percent:
        columns.append(c)

    return df[columns]
```

 ## <mark>Electric and Alternative Fuel Charging Stations - **Transformación y limpieza**

```python
electric_stations = pd.read_csv('/content/drive/MyDrive/Proyecto Final Henry/Datasets/Electric and Alternative Fuel Charging Stations.csv')
print(electric_stations.shape)
electric_stations.head()
```

```python
electric_stations = electric_stations[electric_stations['Fuel Type Code'] == 'ELEC'].reset_index(drop=True)
electric_stations
```

```python
# Veo cuales son las columnas que en su mayoria son valores nulos, para poder eliminarlas ya que no aportan información que aporte valor
electric_stations.isna().sum()
```

Descarto directamente las columnas con un porcentaje de valores nulos mayor que 50%

```python
important_columns = []
for column in electric_stations.columns:
  percent_null = electric_stations[column].isna().sum() * 100 / len(electric_stations)
  if percent_null < 50.0:
    important_columns.append(column)

electric_stations = electric_stations[important_columns]
electric_stations
```

```python
electric_stations.columns
```

```python
electric_stations = electric_stations[['Fuel Type Code',
                                      'Station Name',
                                      'Street Address',
                                      'Country',
                                      'City',
                                      'State',
                                      'ZIP',
                                      'Groups With Access Code',
                                      'Geocode Status',
                                      'Latitude',
                                      'Longitude',
                                      'Updated At',
                                      'Open Date',
                                      'Access Code',
                                      'Restricted Access']]

# Filtro el dataframe para quedarme solo con las estaciones de carga electrica de la ciudad de Nueva York
electric_stations = electric_stations[electric_stations.City == 'New York']
electric_stations.columns
```

### Dataframe complementario de la pagina de [New York State](https://www.nyserda.ny.gov/All-Programs/Drive-Clean-Rebate-For-Electric-Cars-Program/Charging-Options/Electric-Vehicle-Station-Locator#/find/nearest)

```python
electric_sations_ii = pd.read_csv('/content/drive/MyDrive/Proyecto Final Henry/Datasets/Datasets Complementarios/electric_stations_nyc.csv')

# Filtro el dataframe con las mismas columnas del otro dataset, ya que tienen la misma estructura (columnas)
electric_sations_ii = electric_sations_ii[['Fuel Type Code', 'Station Name', 'Street Address', 'Country', 'City',
       'State', 'ZIP', 'Groups With Access Code', 'Geocode Status', 'Latitude',
       'Longitude', 'Updated At', 'Open Date', 'Access Code',
       'Restricted Access']]

# Concateno los dos datasets
full_electric_stations = pd.concat([electric_stations,
                                    electric_sations_ii],
                                    axis=0)

full_electric_stations = full_electric_stations.drop(columns=['Restricted Access']).reset_index(drop=True)
full_electric_stations
```

```python
full_electric_stations = full_electric_stations.drop(columns=['State', 'Country'])
full_electric_stations['Updated At'] = pd.to_datetime(full_electric_stations['Updated At']).dt.date
full_electric_stations['Open Date'] = pd.to_datetime(full_electric_stations['Open Date']).dt.date
full_electric_stations = full_electric_stations.rename(columns={'ZIP':'Postal Code',
                                                                'Groups With Access Code':'Property',
                                                                'Access Code':'Property Code',
                                                                'Updated At':'Update Date'})
full_electric_stations
```
